```python
%python
dbutils.widgets.removeAll()
```

```python
dbutils.widgets.text("catalogo","catalog_pry_final")
dbutils.widgets.text("schema_bronze","bronze")
dbutils.widgets.text("schema_silver","silver")
dbutils.widgets.text("schema_golden","golden")
```

```python
catalogo = dbutils.widgets.get("catalogo")
schema_bronze = dbutils.widgets.get("schema_bronze")
schema_silver = dbutils.widgets.get("schema_silver")
schema_golden = dbutils.widgets.get("schema_golden")
```

```python
lista_bronze_tables = ["departamentos","ordenes","ordenes_prior","pasillo","productos"]
lista_silver_tables = ["dim_productos"]
lista_golden_tables = ["dm_department_performance","dm_top_products","user_profiling"]
```

###Grants Data Engineers

```python
spark.sql(f"""
          GRANT USE CATALOG ON CATALOG {catalogo} TO `data_engineers`
          """)
spark.sql(f"""
          GRANT USE SCHEMA ON SCHEMA {catalogo}.{schema_bronze} TO `data_engineers`
          """)
spark.sql(f"""
          GRANT USE SCHEMA ON SCHEMA {catalogo}.{schema_silver} TO `data_engineers`
          """)
spark.sql(f"""
          GRANT USE SCHEMA ON SCHEMA {catalogo}.{schema_golden} TO `data_engineers`
          """)
spark.sql(f"""
          GRANT CREATE TABLE ON SCHEMA {catalogo}.{schema_bronze} TO `data_engineers`
          """)
spark.sql(f"""
          GRANT CREATE TABLE ON SCHEMA {catalogo}.{schema_silver} TO `data_engineers`
          """)
spark.sql(f"""
          GRANT CREATE TABLE ON SCHEMA {catalogo}.{schema_golden} TO `data_engineers`
          """)
```

```python
for table in lista_bronze_tables:
  spark.sql(f"""
            GRANT SELECT, MODIFY ON TABLE {catalogo}.{schema_bronze}.{table} TO `data_engineers`
            """)
```

```python
spark.sql(f"""
        GRANT SELECT, MODIFY ON TABLE {catalogo}.{schema_silver}.dim_productos TO `data_engineers`
        """)
```

```python
for table in lista_golden_tables:
  spark.sql(f"""
            GRANT SELECT, MODIFY ON TABLE {catalogo}.{schema_golden}.{table} TO `data_engineers`
            """)
```

###Grants Data Custodians

```python
spark.sql(f"""
          GRANT USE CATALOG ON CATALOG {catalogo} TO `data_custodians`
          """)
spark.sql(f"""
          GRANT USE SCHEMA ON SCHEMA {catalogo}.{schema_bronze} TO `data_custodians`
          """)
spark.sql(f"""
          GRANT USE SCHEMA ON SCHEMA {catalogo}.{schema_silver} TO `data_custodians`
          """)
spark.sql(f"""
          GRANT USE SCHEMA ON SCHEMA {catalogo}.{schema_golden} TO `data_custodians`
          """)
```
