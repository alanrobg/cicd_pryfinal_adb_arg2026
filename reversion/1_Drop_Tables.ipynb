```sql
%python
dbutils.widgets.removeAll()
```

```sql
%python
dbutils.widgets.text("container_bronze","bronze")
dbutils.widgets.text("container_silver","silver")
dbutils.widgets.text("container_golden","golden")
dbutils.widgets.text("nameStorage","adlsproybuckd01")
dbutils.widgets.text("catalogo","catalog_pry_final")
```

```sql
%python

container_bronze = dbutils.widgets.get("container_bronze")
container_silver = dbutils.widgets.get("container_silver")
container_golden = dbutils.widgets.get("container_golden")
nameStorage = dbutils.widgets.get("nameStorage")
catalogo = dbutils.widgets.get("catalogo")

ruta_bronze = f"abfss://{container_bronze}@{nameStorage}.dfs.core.windows.net"
ruta_silver = f"abfss://{container_silver}@{nameStorage}.dfs.core.windows.net"
ruta_golden  = f"abfss://{container_golden}@{nameStorage}.dfs.core.windows.net"
```

### Eliminacion tablas Bronze

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.bronze.departamentos;
DROP TABLE IF EXISTS ${catalogo}.bronze.ordenes;
DROP TABLE IF EXISTS ${catalogo}.bronze.ordenes_prior;
DROP TABLE IF EXISTS ${catalogo}.bronze.pasillo;
DROP TABLE IF EXISTS ${catalogo}.bronze.productos;
```

```sql
%python
## REMOVE DATA (Bronze)
dbutils.fs.rm(f"{ruta_bronze}/tablas/departamentos", True)
dbutils.fs.rm(f"{ruta_bronze}/tablas/ordenes", True)
dbutils.fs.rm(f"{ruta_bronze}/tablas/ordenes_prior", True)
dbutils.fs.rm(f"{ruta_bronze}/tablas/pasillo", True)
dbutils.fs.rm(f"{ruta_bronze}/tablas/productos", True)
```

### Eliminacion tablas Silver

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.silver.dim_productos;
```

```sql
%python
# REMOVE DATA (Silver)
dbutils.fs.rm(f"{ruta_silver}/tablas/dim_productos", True)
```

### Eliminacion tablas Golden

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.golden.dm_department_performance;
DROP TABLE IF EXISTS ${catalogo}.golden.dm_top_products;
DROP TABLE IF EXISTS ${catalogo}.golden.user_profiling;
```

```sql
%python
# REMOVE DATA (Golden)
dbutils.fs.rm(f"{ruta_golden}/tablas/dm_department_performance", True)
dbutils.fs.rm(f"{ruta_golden}/tablas/dm_top_products", True)
dbutils.fs.rm(f"{ruta_golden}/tablas/user_profiling", True)
```
